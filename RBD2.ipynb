```python
import pandas as pd
import math
from scipy import stats
import scipy
import statsmodels.api as sm
from statsmodels.formula.api import ols
from matplotlib import pyplot as plt
```

```python
can={'S1':[75,74,70,73,76,73],'S2':[76,74,71,72,73,73],'S3':[78,74,75,77,76,73]}
data=pd.DataFrame(can)
data
```

```python
data_new=pd.melt(data.reset_index(),id_vars=['index'],value_vars=['S1','S2','S3'])
data_new.columns=['Blocks','Treatments','Value']
data_new
```

```python
model=ols('Value ~ C(Treatments)',data=data_new).fit()
anova_table=sm.stats.anova_lm(model,typ=1)
anova_table
```

```python
model=ols('Value ~ C(Blocks) + C(Treatments)',data=data_new).fit()
anova_table=sm.stats.anova_lm(model,typ=1)
anova_table
```
