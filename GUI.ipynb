```python
from tkinter import *
import sqlite3
```

```python
root=Tk()
root.geometry("500x500")
root.title("Registration Form")

Fullname=StringVar()
Email=StringVar()
var=IntVar()
c=StringVar()
var1=IntVar()
var2=IntVar()
var3=IntVar()

def database():
    name1=Fullname.get()
    email=Email.get()
    gender=var.get()
    if gender==3:
        s="MALE"
    else:
        s="FEMALE"
    country=c.get()
    prog1=var1.get()
    if prog1==1:
        s1="JAVA"
    else:
        s1=""
    prog2=var2.get()
    if prog2==1:
        s2="C++"
    else:
        s2=""
    prog3=var3.get()
    if prog3==1:
        s3="PYTHON"
    else:
        s3=""
    con=sqlite3.connect("Back1.db")
    with con:
        cursor=con.cursor()
    cursor.execute('CREATE TABLE IF NOT EXISTS Student (Fullname TEXT,Email TEXT,Gender TEXT,country TEXT,Programming TEXT)')
    cursor.execute('INSERT INTO Student (Fullname,Email,Gender,country,Programming) VALUES(?,?,?,?,?)',(name1,email,s,country,s1,))
    con.commit()
def reset():
    Fullname.set("")
    Email.set("")
    var.set("0")
    c.set("Select Your Country")
    var1.set("0")
    var2.set("0")
    var3.set("0")
    var4.set("0")
```

```python
label_0=Label(root,text="Registration Form",width=20,font=("bold",20))
label_0.place(x=90,y=53)

label_1=Label(root,text="Full Name",width=20,font=("bold",10))
label_1.place(x=80,y=130)
entry_1=Entry(root,textvar=Fullname)
entry_1.place(x=240,y=130)

label_2=Label(root,text="Email",width=20,font=("bold",10))
label_2.place(x=68,y=180)
entry_2=Entry(root,textvar=Email)
entry_2.place(x=240,y=180)

label_3=Label(root,text="Gender",width=20,font=("bold",10))
label_3.place(x=70,y=230)
Radiobutton(root,text="Male",padx=5,variable=var,value=3).place(x=235,y=230)
Radiobutton(root,text="Female",padx=5,variable=var,value=4).place(x=290,y=230)

label_4=Label(root,text="Country",width=20,font=("bold",10))
label_4.place(x=70,y=280)
lis=['Canada','India','UK','Nepal','Iceland','South Africa'];
droplist=OptionMenu(root,c,*lis)
droplist.config(width=16)
c.set('Select Your Country')
droplist.place(x=240,y=280)

label_5=Label(root,text="Programming",width=20,font=("bold",10))
label_5.place(x=85,y=330)
Checkbutton(root,text="JAVA",variable=var1).place(x=235,y=330)
Checkbutton(root,text="C",variable=var2).place(x=290,y=330)
Checkbutton(root,text="PYTHON",variable=var3).place(x=345,y=330)

Button(root,text="SUBMIT",width=20,bg='dark blue',fg='white',command=database).place(x=180,y=380)
Button(root,text="RESET",width=20,bg='dark blue',fg='white',command=reset).place(x=180,y=420)
root.mainloop()
```
