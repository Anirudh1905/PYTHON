```python
import pandas as pd
import numpy as np
import math
from scipy import stats
import scipy
import statsmodels.api as sm
from statsmodels.formula.api import ols
from matplotlib import pyplot as plt
```

```python
data=pd.read_excel('cotton weight.xlsx')
data
```

```python
data_new=pd.melt(data.reset_index(),id_vars=['index'],value_vars=['cotwt.15','cotwt.20','cotwt.25','cotwt.30','cotwt.35'])
data_new.columns=['Index','Treatments','Value']
data_new
```

```python
model=ols('Value ~ C(Treatments)',data=data_new).fit()
anova_table=sm.stats.anova_lm(model,typ=1)
anova_table
```
