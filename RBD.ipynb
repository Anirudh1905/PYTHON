```python
import pandas as pd
import math
from scipy import stats
import scipy
import statsmodels.api as sm
from statsmodels.formula.api import ols
from matplotlib import pyplot as plt
```

```python
can={'System A':[15,14,10,13,16,13],'System B':[15,14,11,12,13,13],'System C':[18,14,15,17,16,13]}
data=pd.DataFrame(can)
data
```

```python
data_new=pd.melt(data.reset_index(),id_vars=['index'],value_vars=['System A','System B','System C'])
data_new.columns=['Blocks','Treatments','Value']
data_new
```

```python
model=ols('Value ~ C(Treatments)',data=data_new).fit()
anova_table=sm.stats.anova_lm(model,typ=1)
anova_table
```

```python
model=ols('Value ~ C(Blocks) + C(Treatments)',data=data_new).fit()
anova_table=sm.stats.anova_lm(model,typ=1)
anova_table
```
