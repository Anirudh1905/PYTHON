```
a=list(range(1,7))
b=[4,6,12,3,2,1]
c=[4,2,8,3,2,1]
e=list(map(lambda x,y,z:(x+y)*z,a,b,c))
print(e)
```

```
list1 = [21, 24, 12, 34, 10, 15,41]
e=list(filter(lambda a:a%4==0,list1))
print(e)
```

```
from functools import reduce
c=[10, 24, 34, 42, 19]
s=reduce(lambda a,b:a+b,c)
t=reduce(lambda a,b:a*b,c)
print(s,t)
```

```
a=list(range(1,7))
b=[4,6,12,3,2,1]
x=list(map(lambda p,q:(p+q),a,b))
y=list(filter(lambda o:o%2==0,x))
print(x,y)
```
