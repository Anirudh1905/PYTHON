```python
import pandas as pd
import math
from scipy import stats
import scipy
import statsmodels.api as sm
from statsmodels.formula.api import ols
from matplotlib import pyplot as plt
```

```python
can={'canteen 1':[4,6,7,2,5],'canteen 2':[3,2,1,7,6],'canteen 3':[5,8,2,2,2]}
data=pd.DataFrame(can)
data
```

```python
data_new=pd.melt(data.reset_index(),id_vars=['index'],value_vars=['canteen 1','canteen 2','canteen 3'])
data_new.columns=['Index','Treatments','Value']
data_new
```

```python
model=ols('Value ~ C(Treatments)',data=data_new).fit()
anova_table=sm.stats.anova_lm(model,typ=1)
anova_table
```
