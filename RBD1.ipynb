```python
import pandas as pd
import math
from scipy import stats
import scipy
import statsmodels.api as sm
from statsmodels.formula.api import ols
from matplotlib import pyplot as plt
```

```python
can={'Chem1':[1.3,1.6,0.5,1.2,1.1],'Chem2':[2.2,2.4,0.4,2.0,1.8],'Chem3':[1.8,1.7,0.6,1.5,1.3],'Chem4':[3.9,4.4,2.0,4.1,3.4]}
data=pd.DataFrame(can)
data
```

```python
data_new=pd.melt(data.reset_index(),id_vars=['index'],value_vars=['Chem1','Chem2','Chem3','Chem4'])
data_new.columns=['Blocks','Treatments','Value']
data_new
```

```python
model=ols('Value ~ C(Treatments)',data=data_new).fit()
anova_table=sm.stats.anova_lm(model,typ=1)
anova_table
```

```python
model=ols('Value ~ C(Blocks) + C(Treatments)',data=data_new).fit()
anova_table=sm.stats.anova_lm(model,typ=1)
anova_table
```
